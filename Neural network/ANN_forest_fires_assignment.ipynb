```python
X = data.loc[:, data.columns !="size_category"]
Y = data["size_category"]
```

```python
X
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state =0)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
x_test.shape
```

```python
y_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
x_train
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import LeakyReLU,PReLU,ELU
from tensorflow.keras.layers import Dropout
```

```python
# Initialising the ANN
classifier = Sequential()
```

```python
# Adding the input layer 
classifier.add(Dense(units=27,activation='relu'))
```

```python
#the first hidden layer
classifier.add(Dense(units=7,activation='relu'))
```

```python
#he second hidden layer
classifier.add(Dense(units=6,activation='relu'))
```

```python
#the output layer
classifier.add(Dense(units=1,activation='sigmoid'))
```

```python
classifier.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
#Early stopping

import tensorflow as tf


early_stop = tf.keras.callbacks.EarlyStopping(
    monitor="val_loss",
    min_delta=0.001,
    patience=20,
    verbose=1,
    mode="auto",
    baseline=None,
    restore_best_weights=False,
)
```

```python
model_history=classifier.fit(x_train,y_train,validation_split=0.33,batch_size=10,epochs=150 , callbacks=early_stop)
```

```python
model_history.history.keys()
```

```python
# Predicting the Test set results

y_pred = classifier.predict(x_test)
y_pred = (y_pred > 0.5)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
# Calculate the Accuracy
from sklearn.metrics import accuracy_score
score=accuracy_score(y_pred,y_test)
```
