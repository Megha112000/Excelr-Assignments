```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import GridSearchCV
```

```python
RF = RandomForestClassifier()
```

```python
params = {'criterion':['gini','entropy'],'max_depth': np.arange(3, 15),'n_estimators' : np.arange(10,300,50)}
```

```python
gcv = GridSearchCV(RF, params, cv=5)
```

```python
import warnings
warnings.filterwarnings('ignore')

gcv.fit(x,y)
```

```python
gcv.best_params_
```

```python
gcv.best_score_
```

```python
rfc =RandomForestClassifier(n_estimators= 210, max_depth=14, criterion='entropy', random_state=100)
```
